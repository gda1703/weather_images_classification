### 4.2 Data augmentation

La **Data augmentation** est une technique qui permet de générer des images pour augmenter et varier un jeu de données et cela permet d''améliorer la performance d'un modèle de Deep learning. Dans le cadre des images, la génération de nouvelles données peut être faite en appliquant des transformations géométriques sur les images (zoom, rotation, redimensionnement, changement de la luminosité ...) originales.

La fonction **ImageDataGenerator** de la classe keras.preprocessing.image génère des lots de données d’images avec une augmentation des données en temps réel (transformation). Dans notre cas, nous avons choisit les arguements suivants : 
- **rescale** : permet de normaliser les images 
- **rotation_range** : répresente la plage de degrés pour la rotation aléatoire de l'image
- **shear_range** : répresente la plage de degrés pour l'angle d'étirement de l'image
- **zoom_range** : répresente la plage de zoom aléatoire.
- **horizontal_flip** : si **True** retourne aléatoirement l'image horizontalement.

Après la transformation des images, nous avons créer un itérateur permettant de générer un nouveau lot de données à chaque itération pendant l'apprentissage. Nous avons donc utilisé la méthode **.flow_from_directory()** de notre générateur d'images. Dans notre cas, elle prend comme argument :
- **directory** : chemin du répertoire des images (par défaut: aucun)
- **class_mode** : le mode d'acception de l'ensemble des classes. Pour notre problème nous avons choisit **categorical** qui est un tableau numpy de 2D du one-hot encoding des classes
- **target_size** : représente la taille (hauteur, largeur) du redimensionnement des images.
- **batch_size** : représente la taille du jeu de données généré à chaque itération.

Pour conclure, les générateurs de données permettent de charger les images lorsque le réseau en a besoin, et de leur appliquer des transformations pour chaque itération. Cette approche permet d'augmenter notre jeu de données et d'éviter au maximum le sur-apprentissage.

```python
# Data augmentation sur le jeu de données de train
train_datagen=ImageDataGenerator(rescale=1./255,
    rotation_range=30,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_width , img_height),
    batch_size= batch_size,
    class_mode='categorical')
```

```python
# Data augmentation sur le jeu de données de validation
val_datagen=ImageDataGenerator(rescale=1./255)

validation_generator=val_datagen.flow_from_directory(
    val_dir,
    target_size=(img_width , img_height),
    batch_size= batch_size,
    class_mode='categorical')
```

#### 5.1. LaNet 

L’architecture **LeNet** a été introduite par Yann LeCun en Novembre 1998 dans le journal Proceedings of the IEEE. C'est l'une des premières et plus simples architectures de réseau de neurones convolutifs (CNN). Au départ, elle a été construite pour la reconnaissance de caratères manuscrits et a ouvert la voie à des nombreuses autres architectures de CNN plus complexes. Elle est donc devenue le point de départ pour la classification d'images.

L'architecture du LeNet est constituée des couches suivantes:
- Convolution 1 : 30 filtres, dimension d'entrée (28, 28, 1), dimension du noyau (5, 5), fonction d'activation ReLU, pas de dépassement du noyau. Dans notre cas, les images sont de la taille 224x224 donc nous avons changé la dimension d'entrée pour notre problème.
- Max-Pooling 1 : dimension du pooling (2, 2).
- Convolution 2 : 16 filtres, dimension du noyau (3, 3), fonction d'activation ReLU, Pas de dépassement du noyau.
- Max-Pooling 2 : dimension du pooling (2, 2).
- Dropout : Connexions coupées: 20%. (que nous avons mis à 40%)
- Aplatissement
- Dense 1 : 128 neurones, fonction d'activation ReLU.
- Dense 2 : 10 neurones, fonction d'activation softmax (que nous avons mis à 4 puisque nous avons 4 labels)

L'architecture **LeNet est une des plus simples** à implémenter et une des plus importantes en Deep Learning car elle est le pilier de nombreux modèles plus complexe.

#### 5.1.1. Construction du modèle

```python
inputs=Input(shape=(img_width, img_height, 3))

conv_1=Conv2D(30, kernel_size=(5,5), padding='valid', input_shape=(img_width, img_height, 3),  activation='relu')
pooling_1=MaxPooling2D(pool_size=(2, 2))
conv_2=Conv2D(16, kernel_size=(3,3), padding='valid', activation='relu')
pooling_2=MaxPooling2D(pool_size=(2, 2))
dropout=Dropout(rate=0.4)
flatten=Flatten()

dense_1=Dense(units=128, activation='relu')
dense_2=Dense(units=4, activation='softmax')

x=conv_1(inputs)
x=pooling_1(x)
x=conv_2(x)
x=pooling_2(x)

x=dropout(x)
x=flatten(x)
x=dense_1(x)
outputs=dense_2(x)

lenet = Model(inputs = inputs, outputs = outputs)
```

```python
# Compilation du modèle
optimizer = Adam(learning_rate=0.00001)
lenet.compile(loss='categorical_crossentropy', #fonction de perte 
              optimizer=optimizer,             #algorithme de descente de gradient
              metrics='accuracy')              #métrique d'evaluation
```

```python
# Resumé du modèle
lenet.summary()
```

```python
# Entrainement du modèle 
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

history_lenet = lenet.fit(train_generator, 
                              validation_data=validation_generator,  epochs=25, callbacks=[early_stop])  
```

Le nombre d'époques que nous avons mis est 25, mais comme nous pouvons le voir ci-dessus, le processus d'entrainement s'est arrêté après **5 époques**. Cela indique que la condition d'arrêt précoce a été remplie à la 5ème époque, ce qui signifie que les performances du modèle sur l'ensemble de validation ne se sont pas améliorées pendant un certain nombre d'époques (cela permet d'économiser les ressources de calcul et d'éviter le surajustement).

#### 5.1.3. Prévisions

```python
# Traitement des images de test pour pouvoir faire la prévision
resized_test_files = []
for img in test_files:
    img = img.resize((224, 224))
    resized_test_files.append(img)

test_array = np.array([np.array(img) for img in resized_test_files])
test_array= test_array/255.0

# Prédictions 
predicted_probabilities = lenet.predict(test_array)
predicted_labels = np.argmax(predicted_probabilities, axis=1)
```

```python
# Conversion des labels en entier 
le = LabelEncoder()
test_labels_encoded = le.fit_transform(test_labels)

# Matrice de confusion
num_classes = len(le.classes_)
confusion_matrix = np.zeros((num_classes, num_classes), dtype=int)

for true_label, predicted_label in zip(test_labels_encoded, predicted_labels):
    confusion_matrix[true_label, predicted_label] += 1
```

```python
# Création et affichage de la matrice de confusion 
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(test_labels_encoded, predicted_labels)

# Définition des labels 
labels = ['Cloudy', 'Rain', 'Shine', 'Sunrise']

# Affichage de la matrice de confusion
sns.heatmap(cm, annot=True, cmap='Blues', fmt='d', xticklabels=labels, yticklabels=labels)
plt.xlabel('Label prédit')
plt.ylabel('Label réel')
plt.title("Matrice de confusion LaNet")
plt.show()
```

La **matrice de confusion** ci-dessus nous permet de résumer les performances du modèle, en comparant les labels prédits avec les labels réelles. La diagonale de la matrice représente les images correctement classées, tandis que les éléments hors diagonale représentent les erreurs de classification. En colonne nous pouvons voir les labels prédits et en ligne ceux réels.

Comme nous pouvons le voir, le modèle n’arrive pas à prédire à 100% car il y a des images qui se situent en dehors de la diagonale. En prenant un exemple, pour le label **Cloudy**, nous pouvons constater que 23 images sur 30 sont bien classées, cependant le modèle classe quatre images en **Rain** et trois images en **Shine**. Le même raisonnement se poursuit pour les labels restants.

#### 5.2.1. Création du modèle

**InceptionV3** est un modèle de réseau de neurones profond pour la classification d'images mais aussi pour les véhicules autonomes et l'imagerie médicale et il a été développé par des chercheurs de Google en 2015 en tant qu'amélioration par rapport à son prédécesseur, InceptionV1.

Le modèle InceptionV3 utilise une architecture de réseau de neurones convolutif avec plusieurs couches, y compris des couches convolutives, des couches de regroupement et des couches entièrement connectées. L'une de ses principales caractéristiques est l'utilisation de **modules de démarrage**, qui sont conçus pour capturer efficacement des caractéristiques à différentes échelles et résolutions.
Le modèle est formé sur un grand ensemble de données d'images labellisées et apprend à reconnaître les modèles et les caractéristiques dans les images. Une fois formé, le modèle peut être utilisé pour classer de nouvelles images dans différentes catégories ou détecter des objets dans des images.

**InceptionV3** a atteint des performances de pointe sur une gamme d'ensembles de données de classification d'images de référence, tels qu'ImageNet.

Dans un premier temps, nous avons chargé le modèle InceptionV3 pre-entrainé qui a été formé sur le jeu de données ImageNet. Les trois paramentrès que nous avons sont : 
- weights spécifie que les poids pré-formés doivent être chargés.
- include_top = False pour exclure la dernière couche entièrement connectée du modèle, ce qui nous permet d'ajouter nos propres couches de classification. 
- input_shape spécifie la taille des images d'entrée attendues par le modèle.


Ensuite, nous avons crée un modèle sequentiel auquel nous avons ajouté le modèle InceptionV3 en tant que première couche. Une couche **GlobalAveragePooling2D** est ensuite ajoutée pour réduire les dimensions spatiales de la sortie du modèle pré-entraîné.  Puis une couche **Dense** entièrement connectée avec 250 unités et une fonction d'activation ReLU, une couche **Dropout** pour régulariser le réseau mise à 0.4 (cela signifie que le 40% les connextion entre les couche précédente et la couche suivant ont été coupées), et une couche **Dense finale** avec 4 unités qui correspondent au nombre des classes d'output avec une fonction d'activation softmax pour renvoyer des prédictions de probabilité pour chaque classe.

Pour la compilation du modèle, nous avons utilisé l'optimiseur **Adam** avec un taux d'apprentissage de 0,0001 et la fonction de perte d'entropie croisée catégorielle est utilisée pour former le modèle. 

Concernant l'entrainement du modèle, le modèle a été entrainé sur les données d'entraînement (train_generator) et validé sur les données de validation (validation_generator) pour un maximum de 25 époques. Un rappel **EarlyStopping** est également utilisé pour arrêter l'entraînement plus tôt si la perte de validation ne s'améliore pas pendant 3 époques consécutives. Cela permet d'optimiser le taux d'apprentissage du modèle. 

```python
# Modèle InceptionV3 pre-entrainé
base_model_inceptionv3 = InceptionV3(include_top=False, input_shape=(img_width, img_height,3))
```

```python
# Freeze des poids
base_model_inceptionv3.trainable = False
```

```python
# Architecture du modèle
"""model_inception = Sequential([
    base_model_inceptionv3,
    GlobalAveragePooling2D(),
    Dense(250, activation='relu'),
    Dropout(0.4),
    Dense(4, activation='softmax')])"""

model_inception = Sequential()
model_inception.add(base_model_inceptionv3)
model_inception.add(GlobalAveragePooling2D())
model_inception.add(Dense(250, activation='relu'))
model_inception.add(Dropout(rate=0.4))
model_inception.add(Dense(4, activation='softmax')) 
```

```python
# Resumé du modèle
model_inception.summary()
```

```python
# Entrainement du modèle
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

history_inception = model_inception.fit(train_generator, 
                              validation_data=validation_generator, epochs=25,
                                        callbacks=[early_stop])
```

### 5.3 EfficientNet
---

#### 5.3.1. Creation du modèle

**EfficientNet** est une architecture de réseau de neurones profonds qui a été développée en 2019. Ce modèle fait partie des modèles du **state-of-the-art** des performances sur les différentes tâches de Computer Vision.

L'architecture d'EfficientNet est basée sur une méthode de mise à l'échelle composée qui permet au modèle d'équilibrer sa profondeur, sa largeur et sa résolution à chaque couche. Plus précisément, le modèle est composé de plusieurs blocs (qui contient une série de couches convolutives, de normalisation par lots et de fonctions d'activation).

Le bloc de construction principal de l'architecture EfficientNet est la convolution de goulot d'étranglement inversé mobile (**MBConv**). Le bloc MBConv se compose de trois composants principaux :

- Convolution en profondeur : il s'agit d'une opération de convolution légère qui applique un seul filtre à chaque canal d'entrée séparément. Elle est suivie d'une normalisation et d'une activation par lots.

- Goulot d'étranglement inversé : il s'agit d'une fonction non linéaire qui augmente le nombre de canaux, puis applique une convolution ponctuelle pour réduire la dimensionnalité.

- Goulot d'étranglement linéaire : il s'agit d'une combinaison de fonctions de convolution ponctuelle, de normalisation par lots et d'activation qui réduisent le nombre de canaux à la valeur d'origine.

L'architecture EfficientNet comprend également une technique appelée **activation swish**, qui est une fonction d'activation non linéaire qui est plus efficace en termes de calcul que l'activation ReLU.

Enfin, EfficientNet utilise une technique appelée **mise à l'échelle composée**, qui consiste à mettre à l'échelle la profondeur, la largeur et la résolution du modèle de manière équilibrée. Cela permet au modèle d'atteindre une grande précision sur un large éventail de tâches tout en étant efficace en termes de calcul.

Concernant la construction de l'architecture, nous avons procedé comme pour le modèle InceptionV3.

```python
# Modèle EfficientNet
base_model_efn = efn.EfficientNetB0(input_shape=(img_width, img_height, 3), include_top = False, weights = 'imagenet')
```

```python
# Freeze les poids du modèle
for layer in base_model_efn.layers:
    layer.trainable = False
```

```python
# Architecture du modèle 
model_efn = Sequential()
model_efn.add(base_model_efn)
model_efn.add(GlobalAveragePooling2D())
model_efn.add(Dense(250, activation='relu'))
model_efn.add(Dropout(rate=0.4))
model_efn.add(Dense(4, activation='softmax')) 
```

```python
# Compilation du modèle 
optimizer = Adam(learning_rate=0.0001)
model_efn.compile(loss='categorical_crossentropy', #fonction de perte 
              optimizer=optimizer,             #algorithme de descente de gradient
              metrics='accuracy')              #métrique d'evaluation
```

```python
# Resumé du modèle 
model_efn.summary()
```

```python
# Entrainement du modèle
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

history_efn = model_efn.fit(train_generator, 
                              validation_data=validation_generator,  epochs=25, 
                                callbacks=[early_stop])
```

Le nombre d'époques que nous avons mis est 25, mais comme nous pouvons le voir ci-dessus, le processus d'entrainement s'est arrêté après **8 époques**. Cela indique que la condition d'arrêt précoce a été remplie à la 8emè époque, ce qui signifie que les performances du modèle sur l'ensemble de validation ne se sont pas améliorées pendant un certain nombre d'époques (cela permet d'économiser les ressources de calcul et d'éviter le surajustement).

```python
# Prédictions 
predicted_probabilities_efn = model_efn.predict(test_array)
predicted_labels_efn = np.argmax(predicted_probabilities_efn, axis=1)
```

```python
# Creation et affichage de la matrice de confusion 
# Conversion des labels en entier 
le = LabelEncoder()
test_labels_encoded = le.fit_transform(test_labels)

from sklearn.metrics import confusion_matrix

cm_inc = confusion_matrix(test_labels_encoded, predicted_labels_efn)

# Définition des labels 
labels = ['Cloudy', 'Rain', 'Shine', 'Sunrise']

# Affichage de la matrice de confusion
sns.heatmap(cm_inc, annot=True, cmap='Blues', fmt='d', xticklabels=labels, yticklabels=labels)
plt.xlabel('Label prédit')
plt.ylabel('Label réel')
plt.title("Matrice de confusion EfficientNet")
plt.show()
```

Comme nous pouvons le voir, le modèle n’arrive pas à prédire à 100% car il y a des images qui se situent en dehors de la diagonale. En prenant un exemple, pour le label **Shine**, nous pouvons constater que 22 images sur 26 sont bien classées, cependant le modèle classe deux images comme **Cloudy** et deux images comme **Sunrise**. Le même raisonnement se poursuit pour les labels restants. 

L'erreur de la classification entre Shine et Sunrise pourrait etre du au fait que dans les 2 types de labels, il y a le soleil et que parfois le levé du soleil (sunrise) pourrait etre déjà haut et donc il y a plus de lumière du soleil ce qui peut faire penser au label shine.
